# Exemplo de c√≥digo - Diagrama de Bode

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Defini√ß√£o da fun√ß√£o de transfer√™ncia do sistema
num = [1]  # Numerador
den = [1, 2, 1]  # Denominador
sys = signal.TransferFunction(num, den)

# Frequ√™ncias para avalia√ß√£o do sistema
frequencies = np.logspace(-1, 2, num=1000)

# C√°lculo da resposta em frequ√™ncia
w, mag, phase = signal.bode(sys, frequencies)

# Plot do gr√°fico de amplitude
plt.figure()
plt.semilogx(w, mag)  # Escala logar√≠tmica no eixo x
plt.xlabel('Frequ√™ncia [rad/s]')
plt.ylabel('Amplitude [dB]')
plt.title('Diagrama de Bode - Amplitude')
plt.grid(True)
plt.show()

# Plot do gr√°fico de fase
plt.figure()
plt.semilogx(w, phase)  # Escala logar√≠tmica no eixo x
plt.xlabel('Frequ√™ncia [rad/s]')
plt.ylabel('Fase [graus]')
plt.title('Diagrama de Bode - Fase')
plt.grid(True)
plt.show()
```
