# Exemplo de código - Diagrama de Bode

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Definição da função de transferência do sistema
num = [1]  # Numerador
den = [1, 2, 1]  # Denominador
sys = signal.TransferFunction(num, den)

# Frequências para avaliação do sistema
frequencies = np.logspace(-1, 2, num=1000)

# Cálculo da resposta em frequência
w, mag, phase = signal.bode(sys, frequencies)

# Plot do gráfico de amplitude
plt.figure()
plt.semilogx(w, mag)  # Escala logarítmica no eixo x
plt.xlabel('Frequência [rad/s]')
plt.ylabel('Amplitude [dB]')
plt.title('Diagrama de Bode - Amplitude')
plt.grid(True)
plt.show()

# Plot do gráfico de fase
plt.figure()
plt.semilogx(w, phase)  # Escala logarítmica no eixo x
plt.xlabel('Frequência [rad/s]')
plt.ylabel('Fase [graus]')
plt.title('Diagrama de Bode - Fase')
plt.grid(True)
plt.show()
```
