# Exemplo - SÃ©rie de Fourier

```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_square_wave(t, amplitude, frequencia, num_harmonicos):
    onda_quadrada = np.zeros_like(t)

    for n in range(1, num_harmonicos+1, 2):
        coeficiente = (4 / (n * np.pi)) * np.sin(n * 2 * np.pi * frequencia * t)
        onda_quadrada += coeficiente

    return amplitude * onda_quadrada

amplitude = 1  # Amplitude da onda quadrada
frequencia = 1  # FrequÃªncia da onda quadrada
periodo = 1 / frequencia  # PerÃ­odo da onda quadrada
tempo_total = 4 * periodo  # Tempo total da simulaÃ§Ã£o
num_pontos = 1000  # NÃºmero de pontos para a simulaÃ§Ã£o

tempo = np.linspace(0, tempo_total, num_pontos)

# Onda quadrada original
onda_original = amplitude * np.sign(np.sin(2 * np.pi * frequencia * tempo))

# Onda quadrada com diferentes nÃºmeros de harmÃ´nicos
num_harmonicos = [1, 5, 10]
ondas_fourier = []

for num in num_harmonicos:
    onda_fourier = fourier_square_wave(tempo, amplitude, frequencia, num)
    ondas_fourier.append(onda_fourier)

# Plotando as ondas
plt.figure(figsize=(10, 6))

plt.plot(tempo, onda_original, label='Onda Quadrada Original')

for i, num in enumerate(num_harmonicos):
    plt.plot(tempo, ondas_fourier[i], label=f'{num} harmÃ´nicos')

plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.title('Onda Quadrada - RepresentaÃ§Ã£o de Fourier com Diferentes NÃºmeros de HarmÃ´nicos')
plt.grid(True)
plt.legend()
plt.show()
```
