# Cálculo da resposta em frequência para um SLIT

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Inicializando listas que armazenarão módulo e fase da FT do SLIT
mod = []
fase = []
w_ = []

# Funçao de transferência utilizada: H(s) = (s+0.1)/(s+5)
for w in np.linspace(0.00001, 50, num=1_001):
    
    # Calculando modulo e fase para vários valores em frequência
    mod.append(np.sqrt(w**2 + 0.01)/np.sqrt(w**2 + 25))
    fase.append(math.atan(w/0.1) - math.atan(w/5))
    w_.append(w)
    
# Realizando o plot da imagem

# Plotando a amplitude, com destaque em 2 rad/s
mod_2 = np.sqrt(2**2 + 0.01)/np.sqrt(2**2 + 25) # Módulo de H(j2) 
plt.figure()
plt.plot(w_,mod)
plt.plot(2,mod_2,'.r') 
plt.xlabel('Frequência (rad/s)')
plt.ylabel('|$H(j\omega)$|')
plt.grid()
plt.legend(['Resposta em frequência - Módulo'])

# Plotando a fase, com destaque em 2 rad/s
fase_2 = math.atan(2/0.1) - math.atan(2/5)
plt.figure()
plt.plot(w_,fase)
plt.plot(2,fase_2,'r.')
plt.xlabel('Frequência (rad/s)')
plt.ylabel('Fase ($H(j\omega)$)')
plt.grid()
plt.legend(['Resposta em frequência - Fase'])

# Módulo e fase para w = 2rad/s
print(f'\nMódulo de H(j2): {mod_2}')
print(f'\nFase de H(j2): {mod_2}')
```
